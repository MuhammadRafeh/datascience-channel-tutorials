```python
df = pd.read_excel("Customer Call List.xlsx", na_values="N/a")
df
```

```python
df.info()
```

```python
df['CustomerID'].duplicated().any()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop('Not_Useful_Column', axis=1, inplace=True)
df
```

```python
df.columns = ['_'.join(col.lower().strip().split(' ')) for col in df.columns]
```

```python
df
```

```python
df['last_name'].unique()
```

```python
df['last_name'] = df['last_name'].str.strip('/._')
```

```python
df
```

```python
df['phone_number'].unique()
```

```python
# df['phone_number'].str.replace('[^\d]', '', regex=True)
df['phone_number'] = df['phone_number'].apply(lambda x: str(x) if isinstance(x, int) else x)
```

```python
df
```

```python
df['phone_number'] = df['phone_number'].str.replace('[^\d]', '', regex=True)
```

```python
df
```

```python
df['phone_number'] = df['phone_number'].apply(lambda x: f'{x[0:3]}-{x[3:6]}-{x[6:10]}' if isinstance(x, str) else x)
```

```python
df
```

```python
df.dropna(subset='phone_number', inplace=True)
```

```python
df
```

```python
df[['street', 'state', 'zip_code']] = df['address'].str.split(',', n=2, expand=True)
```

```python
df
```

```python
df.drop('address', axis=1, inplace=True)
```

```python
df
```

```python
df['paying_customer'] = df['paying_customer'].str.replace('Yes', 'Y')
df['paying_customer'] = df['paying_customer'].str.replace('No', 'N')
```

```python
df
```

```python
df['do_not_contact'] = df['do_not_contact'].str.replace('Yes', 'Y')
df['do_not_contact'] = df['do_not_contact'].str.replace('No', 'N')
```

```python
df
```

```python
df['do_not_contact'] = df['do_not_contact'].fillna('N')
```

```python
df
```

```python
df.fillna('', inplace=True)
```

```python
df
```

```python
df.drop(df[df['do_not_contact'] == 'Y'].index, inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```
