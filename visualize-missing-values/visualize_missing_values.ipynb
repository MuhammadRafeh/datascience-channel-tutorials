```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
data = {'A': [1,2,None,4,5],
       'B': [None,2,3,4,None],
       'C':[1,2,3,4,5]}
df = pd.DataFrame(data)
df.head()
```

```python
# Method 1
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values')
plt.xlabel('Column')
```

```python
# Method 2
na_values = df.isnull().sum()
na_values.plot(kind='bar')
plt.title('Missing Values')
plt.xlabel('Columns')
plt.ylabel('Count of null Values')
```

```python
# Method 3
sns.barplot(x=na_values.index, y=na_values)
plt.title('Missing Values')
plt.xlabel('Columns')
plt.ylabel('Count of Nan values')
```

```python
# Method 4
msno.bar(df)
```

```python
# Method 5
msno.matrix(df)
plt.title('Matrix Plot')
```
